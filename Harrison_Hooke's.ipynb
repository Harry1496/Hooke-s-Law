# Error Analysis

```python
import numpy as np

def rule_1(c,A_err):
    q = c * A_err
    return q

def rule_2(c,m,A,errA):
    q = abs(c * m * A**(m-1)) * errA
    return q
    
def rule_3(A_err,B_err):
    q = np.sqrt(A_err**2 + B_err**2)
    return q

def rule_4(Q,A,A_err,m,B,B_err,n):
    q_err = abs(Q)*np.sqrt((m*(A_err/A))**2+(n*(B_err/B))**2)
    return q_err

print("For spring 1 method 2")
print('Error in t = 0.05 seconds')
print('Error in T = .005')
Err_T = rule_2(1,2,4.1,0.005)
print('Error in T^2 = ',Err_T)
c = 4*3.14**2
Err_k = rule_2(c,1,0.31,0.01)
print('Error in k = ',Err_k)


print("For spring 2 method 2")
print('Error in t = 0.05 seconds')
print('Error in T = .005')
Err_T = rule_2(1,2,11.22,0.005)
print('Error in T^2 = ',Err_T)
c = 4*3.14**2
Err_k = rule_2(c,1,0.07,0.02)
print('Error in k = ',Err_k)
```
